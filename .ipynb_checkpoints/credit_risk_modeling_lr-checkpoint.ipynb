<h1>Modeling</h1>

```python
#one hot encoding
cred_num = cr_loan_clean.select_dtypes(exclude=['object'])
cred_str = cr_loan_clean.select_dtypes(include=['object'])

cred_str_onehot = pd.get_dummies(cred_str)

cr_loan_prep = pd.concat([cred_num, cred_str_onehot], axis=1)

print(cr_loan_prep.columns)
```

```python
X = cr_loan_prep[['person_age', 'person_income', 'person_emp_length', 'loan_amnt',
       'loan_int_rate', 'loan_percent_income',
       'cb_person_cred_hist_length', 'person_home_ownership_MORTGAGE',
       'person_home_ownership_OTHER', 'person_home_ownership_OWN',
       'person_home_ownership_RENT', 'loan_intent_DEBTCONSOLIDATION',
       'loan_intent_EDUCATION', 'loan_intent_HOMEIMPROVEMENT',
       'loan_intent_MEDICAL', 'loan_intent_PERSONAL', 'loan_intent_VENTURE',
       'loan_grade_A', 'loan_grade_B', 'loan_grade_C', 'loan_grade_D',
       'loan_grade_E', 'loan_grade_F', 'loan_grade_G',
       'cb_person_default_on_file_N', 'cb_person_default_on_file_Y']]

y = cr_loan_prep[['loan_status']]

X_train, X_test,y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=123)
```

```python
X_train.head()
```

```python
clf_logistic = LogisticRegression(solver='lbfgs').fit(X_train, np.ravel(y_train))

lr_preds = clf_logistic.predict_proba(X_test)

lr_preds_df = pd.DataFrame(lr_preds[:,1][0:5], columns = ['prob_default'])
true_df = y_test.head()

#showing first 5 predictions
print(pd.concat([true_df.reset_index(drop=True),lr_preds_df],axis=1))
```

<h3>Model discrimination and Impact</h3>

<p>Thresholds and confusion matrices</p>

```python
avg_loan_amnt = X_test[['loan_amnt']].values.mean()

lr_preds_df['loan_status'] = lr_preds_df['prob_default'].apply(lambda x: 1 if x > 0.4 else 0)

num_defaults = lr_preds_df['loan_status'].value_counts()[1]

default_recall = precision_recall_fscore_support(y_test,lr_preds_df['loan_status'])[1][1]

print(num_defaults * avg_loan_amnt * (1 - default_recall))
```

```python
thresh = [0.2 , 0.225, 0.25 , 0.275, 0.3  , 0.325, 0.35 , 0.375, 0.4,
       0.425, 0.45, 0.475, 0.5, 0.525, 0.55, 0.575, 0.6, 0.625,
       0.65]

def_recalls = []
nondef_recalls = []
accs = []

for threshold in thresh:
    preds_df['loan_status'] = preds_df['prob_default'].apply(lambda x: 1 if x > threshold else 0)
    def_recalls.append(precision_recall_fscore_support(y_test,preds_df['loan_status'])[1][1])
    nondef_recalls.append(precision_recall_fscore_support(y_test,preds_df['loan_status'])[1][0])
    cnf_mat = confusion_matrix(y_test, preds_df['loan_status'])
    accs.append((cnf_mat[0][0]+cnf_mat[1][1])/cnf_mat.sum())

ticks = np.array(thresh)
```

<h3>Threshold selection</h3>

```python
plt.figure(figsize=[15, 7])
plt.plot(thresh,def_recalls)
plt.plot(thresh,nondef_recalls)
plt.plot(thresh,accs)
plt.xlabel("Probability Threshold")
plt.xticks(ticks)
plt.legend(["Default Recall","Non-default Recall","Model Accuracy"])
plt.show()
```
